# ParallelMonitor (alpha)

`ParallelMonitor` is a thin wrapper around Parallel's [Monitor API](https://docs.parallel.ai/monitor-api/overview) — schedule a Task to run on a recurring cadence and get notified when relevant new content shows up.

> **Alpha:** the Monitor API is still in `/v1alpha`. Shapes may change without notice. The Parallel SDK doesn't yet expose this surface, so this module talks to the API directly via httpx.

## Create a monitor

```python
from langchain_parallel import ParallelMonitor, MonitorWebhook

monitor_client = ParallelMonitor()

created = monitor_client.create(
    query="Latest peer-reviewed papers on net-energy-gain fusion",
    frequency="1d",  # <n><unit> in 1h..30d
    metadata={"label": "smoke-test-fusion"},
    # Optional: a webhook URL to receive new events.
    # webhook=MonitorWebhook(
    #     url="https://example.com/parallel/webhook",
    #     event_types=["new_event_group"],
    # ),
)
monitor_id = created["monitor_id"]
print("created", monitor_id, "-", created["status"])
```

## Retrieve / update / list events

```python
got = monitor_client.retrieve(monitor_id)
print("query:", got["query"][:80])
```

List events the monitor has surfaced. Newly-created monitors typically have none yet, so we just verify the call shape:

```python
events = monitor_client.list_events(monitor_id, lookback_period="7d")
print("events:", len(events.get("events", [])))
```

## Delete the monitor

Always delete a monitor when you're done so you stop receiving events. (Smoke-test cleanup happens here too.)

```python
deleted = monitor_client.delete(monitor_id)
print("deleted:", deleted["status"])
```
